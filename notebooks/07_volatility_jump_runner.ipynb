## **1. Project Setup**

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()

if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

print(f"Project root: {PROJECT_ROOT}")
```

## **2. Imports**

```python
from pathlib import Path

from src.volatility_jumps import (
    find_volatility_jumps_from_csv,
    format_volatility_jump_table,
    print_volatility_jump_results,
)

from src.volatility_jump_exports import (
    save_volatility_jump_result,
    print_exported_jump_paths,
)
```

## **3. Settings**

```python
# ---------------------------------------------------------------------------
# Volatility Jump Runner Settings
# ---------------------------------------------------------------------------

CSV_PATH = PROJECT_ROOT / "data" / "processed" / "all_sessions" / "us100_cash_M1_all_sessions_2024_01_01_to_2024_12_31.csv"

ROLLING_WINDOW = 100
JUMP_THRESHOLD = 2.0
TOP_N = 10

ANALYSIS_NAME = "all_sessions_2024_volatility_jumps"


# ---------------------------------------------------------------------------
# Output Settings
# ---------------------------------------------------------------------------

SAVE_RESULTS = True
INCLUDE_TIMESTAMP = True
```

## **4. Detect Volatility Jumps**

```python
jumps = find_volatility_jumps_from_csv(
    csv_path=CSV_PATH,
    rolling_window=ROLLING_WINDOW,
    threshold=JUMP_THRESHOLD,
    top_n=TOP_N,
)

print_volatility_jump_results(jumps)
```

## **5. View Results**

```python
friendly_jumps = format_volatility_jump_table(jumps)

friendly_jumps
```

## **6. Check Size**

```python
jumps.shape
```

## **7. Save Results**

```python
if SAVE_RESULTS:
    export_paths = save_volatility_jump_result(
        jump_df=jumps,
        analysis_name=ANALYSIS_NAME,
        include_timestamp=INCLUDE_TIMESTAMP,
    )

    print_exported_jump_paths(export_paths)

else:
    print("SAVE_RESULTS is False. Results were not saved.")
```
